# 📊 Sales Data Cleaning (Python)
This notebook performs data cleaning and preprocessing for sales dataset to prepare it for analysis and dashboarding.

## 📦 Import Library
Importing necessary library for data manipulation and analysis.

```python
import pandas as pd
```

## 📂 Load Dataset
The dataset is loaded using pandas for further processing.

```python
data = pd.read_csv("C:/Users/shahm/OneDrive/Desktop/Python/sales_data_sample.csv", encoding='latin1')
```

## 🔍 Data Overview
Understanding the structure, columns, and data types.

```python
data.head()
data.info()
```

## 🧹 Clean Column Names
Removed unwanted spaces from column names for consistency.

```python
data.columns = data.columns.str.strip()
```

## 🔄 Data Type Conversion
Converted date and numeric columns into proper format.

```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'],errors='coerce')
data['SALES'] = pd.to_numeric(data['SALES'], errors='coerce')
```

```python
data.isnull().sum()
```

## ⚠️ Handling Missing Values
Filled missing values with appropriate placeholders.

```python
data['ADDRESSLINE2'] = data['ADDRESSLINE2'].fillna("Unknown")
data['STATE'] = data['STATE'].fillna("Unknown")
data['TERRITORY'] = data['TERRITORY'].fillna("Unknown")
data['POSTALCODE'] = data['POSTALCODE'].fillna("Unknown")
```

## 🔎 Checking Missing Values & Duplicates
Verified missing values and duplicate records.

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

## 🧮 Feature Engineering
Extracted Year and Month from Order Date.

```python
data['YEAR'] = data['ORDERDATE'].dt.year
data['MONTH'] = data['ORDERDATE'].dt.month
```

```python
data.info()
```

## 💾 Save Cleaned Data
Exported the cleaned dataset for Power BI dashboard usage.

```python
data.to_csv("C:/Users/shahm/OneDrive/Desktop/Python/clean_sales_data.csv", index=False)
```
